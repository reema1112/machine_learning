```python
final_rating=reviews.merge(apps,on="App")
```

```python
x=final_rating["Rating"].value_counts()>200
x
```

```python
y=x[x].index
y
```

```python
final_rating=final_rating[final_rating["Rating"].isin(y)]
```

```python
final_rating.dropna(inplace=True)
```

```python
final_rating["Sentiment"].replace("Positive",1,inplace=True)
final_rating["Sentiment"].replace("Negative",0,inplace=True)
```

```python
final_rating["Sentiment"].replace("Neutral",2,inplace=True)
final_rating
```

```python
final_rating["Rating"].describe()
```

```python
final_rating["Sentiment"].value_counts()
```

```python
final_rating.drop(columns=["Translated_Review"],inplace=True)
```

```python
final_rating.corr()["Rating"]
```

```python
final_rating.shape
```

```python
final_rating.info()
```

```python
final_rating=final_rating[["App","Sentiment","Sentiment_Polarity","Sentiment_Subjectivity","Reviews","Rating"]]
```

```python
final_rating.head()
```

```python
final_rating.loc[final_rating.Sentiment==1,"Rating"]=4.4
final_rating.loc[final_rating.Sentiment==2,"Rating"]=4.1
final_rating.loc[final_rating.Sentiment==0,"Rating"]=3.3
final_rating.tail(15)
```

```python
final_rating.drop_duplicates(["App","Sentiment","Rating"],inplace=True)
final_rating.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X=final_rating.iloc[:,1:-1].values
y=final_rating.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,accuracy_score
y_pred=reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg3=RandomForestRegressor(n_estimators=1)
```

```python
reg3.fit(X_train,y_train)
```

```python
y_pred2=reg3.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```
